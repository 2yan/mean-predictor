```python
def get_data(filenumber = 1):
    data = pd.read_csv(files[filenumber])
    data.columns = data.columns.str.lower()
    data['date'] = data['unix timestamp'].map(lambda x: datetime.fromtimestamp(x/1000))
    data.sort_values('date', inplace = True)
    data.reset_index(inplace = True, drop = True)
    return data
```

```python


timesteps = 25
X = []
y = []

for i in range(len(data)-  timesteps):
    X.append(data.iloc[i:i + timesteps][['volume_x']].values)
    actual = data.iloc[i + timesteps]['true_avg']
    lagged = data.iloc[i + timesteps]['lagged_avg_x']
    y.append((actual - lagged)/lagged)
    
    
X = np.array(X)
y = np.array(y)
```

```python
sea.distplot(y, kde = False)
```

```python

def generate_model(X, y):
    model = kr.models.Sequential()
    model.add(kr.layers.LSTM(32, input_shape=X.shape[1:]))
    model.add(kr.layers.Dense(16))
    model.add(kr.layers.Dense(1))
    optimizer = kr.optimizers.RMSprop(lr=0.01)
    model.compile(loss='mean_absolute_error', optimizer=optimizer)
    return model

#model = generate_model(X,y)
model.fit(X, y,batch_size=7000,epochs=100, verbose=1)
```

```python
def predict(data):
    timesteps = 25


    ans = pd.Series(index = data.index)
    
    for i in range(timesteps, len(data)):
        x = data.iloc[i - timesteps: i][['close', 'volume']].values
        y = model.predict(np.array([x]))
        ans.iloc[i] = y
        
    return ans
```
